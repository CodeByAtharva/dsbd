```python
import seaborn as sns
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
ds=pd.read_csv("dataset_Facebook.csv",sep=";");
```

```python
ds.head()
```

```python
ds.info()
```

```python
ds.describe()
```

```python
ds.mode().iloc[0]
```

```python
ds.isnull().sum()
```

```python
ds.fillna(ds.mode().iloc[0],inplace=True)
#modifies the original dataframe
```

```python
ds.isnull().sum()
```

```python
ds.columns
```

```python
ds["Type"].unique()
```

```python
ds["Type"]=ds["Type"].map({"Photo":0,"Status":1,"Link":2,"Video":3})
```

```python
x=ds.drop("Total Interactions",axis=1)
y=ds["Total Interactions"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=20,random_state=40)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.coef_
```

```python
ytestpredict=model.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse=mean_squared_error(ytest,ytestpredict)
mae=mean_absolute_error(ytest,ytestpredict)
r2=r2_score(ytest,ytestpredict)
```

```python
print("mse :",mse)
print("mae :",mae)
print("r2 score :",r2)
```
